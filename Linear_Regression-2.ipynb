```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random.mtrand import randint
import math as math
```

```python
x_train = np.array(np.arange(0,20, step=1))
y_train = np.array(np.random.randint(5, size=1) * x_train + np.random.randint(10, size=1))
```

```python
plt.scatter(x_train,y_train)
```

```python
# Adding Noise
noise = np.random.normal(0,5, len(x_train))
y_train = y_train + noise


plt.scatter(x_train,y_train)
```

```python
def compute_cost(x,y,w,b):
  m = x.shape[0]
  f_wb = x * w + b
  cost = (1/(2 * m)) * np.sum((f_wb - y) ** 2)
  return cost
```

```python
def compute_gradient(x,y,w,b):
  m = x.shape[0]
  f_wb = x * w + b
  dj_wb = (1/m) * np.sum((f_wb - y) * x)
  dj_db = (1/m) * np.sum((f_wb - y))
  return dj_wb, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, iterations, cost_function, gradient_function):
    J_history = []
    p_history = []
    b = b_in
    w = w_in
    for i in range(iterations):
        dj_dw, dj_db = gradient_function(x, y, w , b)

        b = b - alpha * dj_db
        w = w - alpha * dj_dw
    return w, b
```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000

alpha = 1.0e-2
# run gradient descent
w_final, b_final = gradient_descent(x_train ,y_train, w_init, b_init, alpha, iterations, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
# Test new w and b

x_test = np.array(np.arange(0,20, step=1))
y_test = x_test * w_final + b_final
```
